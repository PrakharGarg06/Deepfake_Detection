```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.models as models
from torchvision import transforms
from PIL import Image
import cv2
import numpy as np
import random
import os
import glob
from mtcnn.mtcnn import MTCNN
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split

# --- 1. Global Setup and Face Detector ---
print("Initializing MTCNN... (This may take a moment)")
detector = MTCNN()
IMG_SIZE = 224
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# --- 2. (Hardened) Video Processing Functions ---
# These will skip corrupted videos instead of freezing

def extract_key_frames(video_path, frames_per_video=10):
    """Extracts a set number of frames uniformly from a video."""
    cap = cv2.VideoCapture(video_path)
    frames = []
    try:
        frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        if frame_count <= 0:
            print(f"Warning: Could not read video or no frames: {os.path.basename(video_path)}")
            return frames
        indices = np.linspace(0, frame_count - 1, frames_per_video, dtype=int)
        for i in indices:
            cap.set(cv2.CAP_PROP_POS_FRAMES, i)
            ret, frame = cap.read()
            if ret:
                frames.append(frame)
    except Exception as e:
        print(f"!!! SKIPPING VIDEO due to read error: {os.path.basename(video_path)}")
    finally:
        cap.release()
    return frames

def detect_and_crop_face(bgr_frame):
    """Detects a face in a BGR frame, crops it, and returns a resized RGB face."""
    try:
        rgb_frame = cv2.cvtColor(bgr_frame, cv2.COLOR_BGR2RGB)
        result = detector.detect_faces(rgb_frame)
        if not result:
            return None
        x, y_, w, h = result[0]['box']
        x, y_ = max(x, 0), max(y_, 0)
        face_bgr = bgr_frame[y_:y_ + h, x:x + w]
        face_rgb = cv2.cvtColor(face_bgr, cv2.COLOR_BGR2RGB)
        return cv2.resize(face_rgb, (IMG_SIZE, IMG_SIZE))
    except Exception:
        return None

# --- 3. Build Video Lists (with CORRECTED Fake Video Path) ---

def build_ff_video_lists(dataset_path, max_videos_per_class=1000, val_split=0.15):
    """Finds all real/fake videos and creates train/val splits."""
    
    # Correct path for 'original' videos
    real_videos_path = os.path.join(dataset_path, "original", "*.mp4")
    real_videos = glob.glob(real_videos_path)
    
    # --- THIS IS THE CORRECTED PART for 'fake' videos ---
    fake_video_paths = []
    manipulation_folders = [
        "Deepfakes", "Face2Face", "FaceSwap", "NeuralTextures", 
        "FaceShifter", "DeepFakeDetection"
    ]
    
    print("Searching for video files...")
    for folder in manipulation_folders:
        # We removed the "/videos" subfolder from the path
        path = os.path.join(dataset_path, folder, "*.mp4") 
        fake_video_paths.extend(glob.glob(path))
    # --- END OF FIX ---

    print(f"Found {len(real_videos)} total REAL videos.")
    print(f"Found {len(fake_video_paths)} total FAKE videos.")
    
    if not real_videos or not fake_video_paths:
        print("!!! CRITICAL ERROR: Could not find REAL or FAKE videos.")
        print("Please double-check your DATASET_BASE_PATH and folder structure.")
        return [], [], [], []
    
    # Limit the number of videos if specified
    if max_videos_per_class is not None:
        random.shuffle(real_videos)
        random.shuffle(fake_video_paths)
        real_videos = real_videos[:max_videos_per_class]
        fake_video_paths = fake_video_paths[:max_videos_per_class]
        print(f"--- Limiting to {len(real_videos)} real and {len(fake_video_paths)} fake videos ---")

    # Create train/validation splits
    train_real, val_real = train_test_split(real_videos, test_size=val_split, random_state=42)
    train_fake, val_fake = train_test_split(fake_video_paths, test_size=val_split, random_state=42)
    
    print(f"--- Using {len(train_real)} real / {len(train_fake)} fake for TRAINING ---")
    print(f"--- Using {len(val_real)} real / {len(val_fake)} fake for VALIDATION ---")
    
    return train_real, train_fake, val_real, val_fake

# --- 4. PyTorch Dataset Class (Processes Videos) ---

class VideoAsImageDataset(Dataset):
    def __init__(self, real_video_paths: list, fake_video_paths: list, transforms=None):
        self.transforms = transforms
        self.processed_faces = []
        self.labels = []
        
        print(f"--- Processing {len(real_video_paths)} real and {len(fake_video_paths)} fake videos ---")

        print("Processing 'Real' (Original) videos...")
        for video_path in tqdm(real_video_paths):
            for frame in extract_key_frames(video_path):
                face = detect_and_crop_face(frame)
                if face is not None:
                    self.processed_faces.append(face)
                    self.labels.append(1.0) # Real = 1

        print("Processing 'Fake' (Manipulated) videos...")
        for video_path in tqdm(fake_video_paths):
            for frame in extract_key_frames(video_path):
                face = detect_and_crop_face(frame)
                if face is not None:
                    self.processed_faces.append(face)
                    self.labels.append(0.0) # Fake = 0

        self.labels = torch.tensor(self.labels, dtype=torch.float32)

    def __len__(self):
        return len(self.processed_faces)

    def __getitem__(self, idx):
        face_np = self.processed_faces[idx]
        label = self.labels[idx]
        face_pil = Image.fromarray(face_np)
        if self.transforms:
            face_tensor = self.transforms(face_pil)
        return face_tensor, label

# --- 5. Data Augmentation and DataLoaders ---

preprocess_train = transforms.Compose([
    transforms.ToTensor(),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(degrees=10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

preprocess_val = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# --- Create Datasets ---
DATASET_BASE_PATH = "/kaggle/input/ff-c23/FaceForensics++_C23"
# Use 1000 "real" and 1000 "fake" videos total.
# This will be split 85% for train and 15% for validation.
train_real, train_fake, val_real, val_fake = build_ff_video_lists(
    DATASET_BASE_PATH, 
    max_videos_per_class=1000
)

print("\nLoading Training Data (This will take time)...")
train_dataset = VideoAsImageDataset(real_video_paths=train_real,
                                    fake_video_paths=train_fake,
                                    transforms=preprocess_train)

print("\nLoading Validation Data (This will also take time)...")
val_dataset = VideoAsImageDataset(real_video_paths=val_real,
                                  fake_video_paths=val_fake,
                                  transforms=preprocess_val)

# --- Create DataLoaders ---
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

print(f"\nTotal training faces: {len(train_dataset)}")
print(f"Total validation faces: {len(val_dataset)}")

# --- 6. Pre-trained Swin Transformer Model ---

def create_swin_transformer_model():
    """Creates a pre-trained Swin Transformer model and adapts it for binary classification."""
    model = models.swin_t(weights=models.Swin_T_Weights.IMAGENET1K_V1)
    
    # Freeze pre-trained layers
    for param in model.parameters():
        param.requires_grad = False
        
    # Replace the final classifier head
    num_features = model.head.in_features
    model.head = nn.Linear(num_features, 1) # Output 1 logit
    
    return model

model = create_swin_transformer_model()
model.to(device)

# --- 7. Loss, Optimizer, and Training Setup ---
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.head.parameters(), lr=1e-4) # Only optimize the new head
scheduler = optim.lr_scheduler.ExponentialLR(optimizer, gamma=0.85)

num_epochs = 20 # Set a max number of epochs
train_loss_history, train_accuracy_history = [], []
val_loss_history, val_accuracy_history = [], []

# --- 8. Training and Validation Functions ---

def get_accuracy(preds, labels):
    predicted_labels = (preds.flatten() > 0)
    return torch.sum(predicted_labels == labels.to(device)) / labels.numel()

def train_one_epoch(epoch):
    model.train()
    running_loss, running_acc = 0.0, 0.0
    pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Training]")
    for inputs, labels in pbar:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels.unsqueeze(-1))
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        running_acc += get_accuracy(outputs, labels)
        pbar.set_postfix(loss=f"{running_loss / (pbar.n + 1):.4f}", acc=f"{running_acc / (pbar.n + 1):.4f}")
    epoch_loss = running_loss / len(train_loader)
    epoch_acc = running_acc / len(train_loader)
    return epoch_loss, epoch_acc.item()

def validate_one_epoch():
    model.eval()
    val_loss, val_acc = 0.0, 0.0
    pbar_val = tqdm(val_loader, desc="[Validating]")
    with torch.no_grad():
        for inputs, labels in pbar_val:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels.unsqueeze(-1))
            val_loss += loss.item()
            val_acc += get_accuracy(outputs, labels)
            pbar_val.set_postfix(loss=f"{val_loss / (pbar_val.n + 1):.4f}", acc=f"{val_acc / (pbar_val.n + 1):.4f}")
    scheduler.step()
    epoch_loss = val_loss / len(val_loader)
    epoch_acc = val_acc / len(val_loader)
    return epoch_loss, epoch_acc.item()

# --- 9. Main Training Loop with Early Stopping ---

print("\n--- Starting Model Training ---")
best_val_loss = float('inf')
patience = 3 # Number of epochs to wait for improvement
epochs_no_improve = 0
best_model_path = "best_swin_model.pt"

for epoch in range(num_epochs):
    train_loss, train_acc = train_one_epoch(epoch)
    val_loss, val_acc = validate_one_epoch()
    
    train_loss_history.append(train_loss)
    train_accuracy_history.append(train_acc)
    val_loss_history.append(val_loss)
    val_accuracy_history.append(val_acc)
    
    print(f"Epoch {epoch+1} Summary: Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}\n")

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        epochs_no_improve = 0
        print(f"🎉 New best model! Saving model with val_loss: {best_val_loss:.4f}")
        torch.save(model.state_dict(), best_model_path)
    else:
        epochs_no_improve += 1
        print(f"No improvement in validation loss. Patience: {epochs_no_improve}/{patience}")

    if epochs_no_improve >= patience:
        print(f"Early stopping triggered after {epoch+1} epochs.")
        break
        
print("--- Training Finished ---")
print(f"Best validation loss achieved: {best_val_loss:.4f}")

# --- 10. Final Evaluation and Visualization ---

print(f"\nLoading the best model from {best_model_path} for final evaluation...")
# Load the best model weights
model.load_state_dict(torch.load(best_model_path))

# --- Plot Accuracy and Loss Curves ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

ax1.plot(train_accuracy_history, label='Train Accuracy', color='b', marker='o', linestyle='--')
ax1.plot(val_accuracy_history, label='Validation Accuracy', color='r', marker='o')
ax1.set_title('Model Accuracy')
ax1.set_ylabel('Accuracy')
ax1.set_xlabel('Epoch')
ax1.legend(loc='best')

ax2.plot(train_loss_history, label='Train Loss', color='b', marker='o', linestyle='--')
ax2.plot(val_loss_history, label='Validation Loss', color='r', marker='o')
ax2.set_title('Model Loss')
ax2.set_ylabel('Loss')
ax2.set_xlabel('Epoch')
ax2.legend(loc='best')

plt.suptitle('Swin Transformer Training Performance', fontsize=18)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# --- Generate Classification Report and Confusion Matrix ---
model.eval()
all_labels = []
all_preds = []
with torch.no_grad():
    for inputs, labels in tqdm(val_loader, desc="Generating Final Predictions"):
        inputs = inputs.to(device)
        outputs = model(inputs)
        preds = (outputs.flatten() > 0).cpu().numpy()
        all_preds.extend(preds)
        all_labels.extend(labels.cpu().numpy())

all_labels = np.array(all_labels).astype(int)
all_preds = np.array(all_preds).astype(int)
class_names = ['Fake (0)', 'Real (1)']

print("\n" + "="*55)
print("Classification Report")
print("="*55)
print(classification_report(all_labels, all_preds, target_names=class_names))

print("\n" + "="*55)
print("Confusion Matrix")
print("="*55)
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(8, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix', fontsize=16)
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```
